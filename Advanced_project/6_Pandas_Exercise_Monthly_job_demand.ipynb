```python
#conda install datasets matplotlib pandas, conda activate python_course
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

```

```python
## EXERCISE - Monthly Job Demand

df_US = df[df['job_country'] == 'United States'].copy()
```

```python
df_US
```

```python
#want to extract month values but the verbage not month number

df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')

df_US
```

```python
#now need to pivot our data to get it into a format that we can plot it 

df_US_pivot = df_US.pivot_table(index = 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')



```

```python
#need to find a way to sort months chronologically

df_US_pivot.reset_index()
df_US_pivot['month_no'] = pd.to_datetime(
    df_US_pivot['job_posted_month'], format = '%B').dt.month #similar to pd.to_datetime(df['job_posted_date'])

df_US_pivot.sort_values('month_no') 
df_US_pivot.set_index('job_posted_month')
df_US_pivot.drop(columns='month_no')
```

```python
df_US_pivot.plot(kind = 'line')
```

```python
#only want top 3 jobs with most counts

top_3 = df_US['job_title_short'].value_counts().head(3)
top_3= top_3.index.tolist() # need to make it to a list to filter when plotting
```

```python
df_US_pivot[top_3].plot(kind = 'line')
plt.ylabel('Job Count')
plt.xlabel ('2023')
plt.title('Monthly Job Posting for Top Data Jobs (US)')
plt.legend()
plt.show()
```
